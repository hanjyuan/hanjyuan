```python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_equation(x, a, b, c):
    return a*x**2 + b*x + c

a1, b1, c1 = -3, 4, 7
a2, b2, c2 = 5, -4, 9


root1_1 = (-b1 + np.sqrt(b1**2 - 4*a1*c1)) / (2*a1)
root2_1 = (-b1 - np.sqrt(b1**2 - 4*a1*c1)) / (2*a1)

root1_2 = (-b2 + np.sqrt(b2**2 - 4*a2*c2)) / (2*a2)
root2_2 = (-b2 - np.sqrt(b2**2 - 4*a2*c2)) / (2*a2)

vertex_x1 = -b1 / (2*a1)
vertex_x2 = -b2 / (2*a2)

vertex_y1 = quadratic_equation(vertex_x1, a1, b1, c1)
vertex_y2 = quadratic_equation(vertex_x2, a2, b2, c2)


x = np.linspace(-10, 10, 100)


y1 = quadratic_equation(x, a1, b1, c1)
y2 = quadratic_equation(x, a2, b2, c2)


plt.plot(x, y1, label='Y = -3X^2 + 4X + 7')
plt.plot(x, y2, label='Y = 5X^2 - 4X + 9')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quadratic Equation')
plt.grid(True)
plt.legend()

plt.annotate(f'Root 1: ({root1_1:.2f}, 0)', xy=(root1_1, 0), xytext=(root1_1, 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Root 2: ({root2_1:.2f}, 0)', xy=(root2_1, 0), xytext=(root2_1, 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Vertex: ({vertex_x1:.2f}, {vertex_y1:.2f})', xy=(vertex_x1, vertex_y1), xytext=(vertex_x1, vertex_y1+10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.annotate(f'Root 1: ({root1_2:.2f}, 0)', xy=(root1_2, 0), xytext=(root1_2, 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Root 2: ({root2_2:.2f}, 0)', xy=(root2_2, 0), xytext=(root2_2, 10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.annotate(f'Vertex: ({vertex_x2:.2f}, {vertex_y2:.2f})', xy=(vertex_x2, vertex_y2), xytext=(vertex_x2, vertex_y2-10),
             arrowprops=dict(facecolor='black', arrowstyle='->'))

plt.show()
```
